```python
import numpy as np
import matplotlib.pyplot as plt
import mdtraj as md
```

```python
traj = md.load_xtc('md_wrapped.xtc', top = 'topol_water-free.gro')
```

```python
# obtain distance between the first and last atom for each time

def end_end(traj):
    # get the first and last atom
    first = traj.xyz[:,0,:]
    last = traj.xyz[:,-1,:]
    # compute the distance between the first and last atom
    dist = np.sqrt(np.sum((first-last)**2, axis=1))
    return dist

dist = end_end(traj)
```

```python
# obtain radius of gyration for each time

rg = md.compute_rg(traj)
```

```python
#plots

plt.plot(traj.time, dist, 'bo-')
plt.title('End to end distance')
plt.xlabel('Time (ps)')
plt.ylabel('Distance (nm)')

plt.savefig('trp-cage end to end distance')
plt.show()
```

```python
plt.plot(traj.time, rg, 'c*-')
plt.title('Radius of gyration')
plt.xlabel('Time (ps)')
plt.ylabel('Distance (nm)')

plt.savefig('trp-cage radius of gyration')
plt.show()
```
