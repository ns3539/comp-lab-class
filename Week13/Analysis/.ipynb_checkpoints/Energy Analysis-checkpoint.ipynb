```python
import numpy as np
import matplotlib.pyplot as plt
```

## Part 2

```python
# plot potential energy vs time for NPT

f = np.loadtxt('NPT-1.ener')

times = f[:, 1]

PE = f[:, 4]

plt.plot(times, PE)
plt.xlabel('Time (fs)')
plt.ylabel('Potential Energy (a.u.)')
plt.title('Potential Energy vs Time - NPT')
plt.savefig('NPT_PEvsTime.png')
plt.show()
```
