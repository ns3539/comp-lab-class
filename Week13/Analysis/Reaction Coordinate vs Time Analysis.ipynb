```python
import numpy as np
import matplotlib.pyplot as plt
```

## Part 3

```python
#plot reaction coordinate vs time

f= np.loadtxt('MONITOR-COLVAR.metadynLog')

times_monitor = f[:,0] # 5 ps simulation where time goes to 5000, so time must be in fs

# collective variable = a variable that that describes our process of interest 

# in this case it is the difference in distance 
#between the C-O bond that we will be breaking and the C-C bond that we will be forming

colvar_monitor = f[:,1] #distance in angstroms

plt.plot(times_monitor, colvar_monitor)
plt.xlabel('Time (fs)')
plt.ylabel('CC-OC Distance (A)')
plt.title('Reaction Coordinate vs Time')
plt.savefig('ReactionCoordinatevsTime.png')
plt.show()
```
