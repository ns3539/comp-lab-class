```python
import numpy as np
import matplotlib.pyplot as plt
import mdtraj as md
```

### Plot 1: Potential energy vs time for energy minimzation

energy should show convergence to a minimum value

```python
f = open('potential.xvg')
lines = f.readlines()

for num, line in enumerate(lines):
    print(num, line)
```

```python
data = lines[24:] #data starts on line 24

data = np.loadtxt(data,unpack=True)
```

```python
plt.plot(data[0], data[1])
plt.title('Energy Minimization - Steepest Descent')
plt.xlabel('Time (ps)')
plt.ylabel('Potential Energy (kJ/mol)')
plt.savefig('Energy Minimization.png')
plt.show()
```

### Plot 5: RMSD vs time

RMSD should converge to indicate stability. can be compared to pure crystal to determine effectiveness of position restraints

```python
f = open('rmsd.xvg')
lines = f.readlines()

for num, line in enumerate(lines):
    print(num, line)
```

```python
data = lines[18:] #data starts on line 18

data = np.loadtxt(data,unpack=True)
```

```python
g = open('rmsd_xtal.xvg')
lines = g.readlines()

for num, line in enumerate(lines):
    print(num, line)
```

```python
data_xtal = lines[18:] #data starts on line 18

data_xtal = np.loadtxt(data_xtal,unpack=True)
```

```python
plt.plot(data[0], data[1], label = 'equilibrated')
plt.plot(data_xtal[0], data_xtal[1], 'r--', label = 'crystal')

plt.title('RMSD of Backbone')
plt.xlabel('Time (ns)')
plt.ylabel('RMSD (nm)')
plt.legend()
plt.savefig('RMSD')
plt.show()
```
